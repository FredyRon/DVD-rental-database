```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Pregunta 1

def evolucion_ventas(datos, variable):
    fig, ax = plt.subplots()
    datos_filtrados = datos[datos['country'].isin(['India', 'China', 'United States'])]
    datos_filtrados.groupby('country').plot(x='payment_date', y=variable, ax=ax)
    plt.title('Ventas Diarias por Pas (' + variable + ')')
    plt.legend(datos_filtrados.groupby('country').groups.keys())
    return ax

df_datos = pd.read_csv('Query_11.csv')
df_datos["payment_date"] = pd.to_datetime(df_datos["payment_date"])
evolucion_ventas(df_datos, 'total_amount')
plt.show()
```
