```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

Query13_df = pd.read_csv("Query_13.csv")
Query13_df.head()
```

```python
resultados = pd.DataFrame(Query13_df, columns=['country', 'total_amount'])

# Imprimir el DataFrame
print(resultados)
```

```python

# Configurar los datos del grfico
paises = resultados['country']
montos_totales = resultados['total_amount']

# Crear el grfico de barras
plt.figure(figsize=(20, 6))
plt.bar(paises, montos_totales)
plt.xlabel('Pas')
plt.ylabel('Monto total')
plt.title('Monto total por pas')

# Rotar las etiquetas del eje x para una mejor visualizacin
plt.xticks(rotation=90)

# Mostrar el grfico
plt.show()
```
